## Scraping Additional Details from the "Details" link

```python
det=pd.DataFrame(columns=details)
for i in range(990,2000):
    mydict=dict.fromkeys(details)
    url=df['Details'][i]
    page = requests.get(url)
    pagetext = page.text
    soup = BeautifulSoup(pagetext,"lxml")
    tables = soup.findAll('table')
    list_of_rows= []
    for t in tables:
        for row in t.findAll('tr'):
            list_of_cells = []
            for cell in row.findAll(["th","td"]):
                text = cell.text
                list_of_cells.append(text.strip())
            list_of_rows.append(list_of_cells)
    for l in list_of_rows:
        if(len(l)!=0):
            if(l[0] in details):
                 mydict[l[0]]=l[1]
    mydict['Annual Turnover']=list_of_rows[-1][1]
    det = det.append(mydict, ignore_index=True)
    print(i)
```

## Cleaning Data

```python
#merging the details table
df1=pd.read_csv('details_1.csv')
df2=pd.read_csv('details_2.csv')
df3=pd.read_csv('details_3.csv')
df4=pd.read_csv('details_4.csv')
df5=pd.read_csv('details_5.csv')
frames=[df1,df2,df3,df4,df5]
data=pd.concat(frames)

#dropping duplicates
data.drop_duplicates(keep='first', inplace=True)
data.reset_index(inplace=True)

data.shape
```

```python
data['Mailling Address']=data['Mailling Address'].astype(str)
data['Factory Address']=data['Factory Address'].astype(str)
data['Factory Type:']=data['Factory Type:'].astype(str)
data['Principal exportable product']=data['Principal exportable product'].astype(str)
data['No. of Employees']=data['No. of Employees'].astype(str)
```

```python
#code to clean the columns
def clean_text(text):
    text=text.replace('\n','')
    text=text.replace('\t','')
    
    return text
def clean_factory_type(text):
    text=text.replace('Type\nPriority','')
    text=text.replace('\n5',',')
    text=text.replace('\n','')
    return text

def clean_employee(text):
    text=text.replace('\n','')
    text=text.replace('\t','')
    text=text.replace('Management','')
    text=text.replace('Male','')
    text=text.replace('Female','')
    text=text.replace('Employee','')
    text=text.replace(' ','')
    text=text.replace('\xa0','')
    return text
```

```python
data['Mailling Address']=data['Mailling Address'].apply(lambda x: clean_text(x))
data['Factory Address']=data['Factory Address'].apply(lambda x: clean_text(x))
data['Factory Type:']=data['Factory Type:'].apply(lambda x: clean_factory_type(x))
data['Principal exportable product']=data['Principal exportable product'].apply(lambda x: clean_text(x))
data['No. of Employees']=data['No. of Employees'].apply(lambda x: clean_employee(x))
```

```python
data=data.drop('Unnamed: 0',axis=1)
data=data.drop('index',axis=1)
data.head()
```
